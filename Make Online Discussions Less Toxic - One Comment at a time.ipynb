```python
import pandas as pd
import numpy as np
from pandas import read_csv
import os
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
filepath = '/Users/DGS-Mkt/Google Drive/Data Incubator/train.csv'
messages = read_csv(filepath) 
messages.shape
```

```python
#Add column for length of texts
messages['comm_length'] = messages['comment_text'].map(len)
```

```python
messages['comm_length'].plot(bins=200, kind='hist',figsize=(12,5), title='Distributin of Length of Comments in Dataset')
```

```python
#Using data from the labeled columns, add a column that shows 
min_text = messages.iloc[:,:8]
sum_toxic_scores = min_text.iloc[:,2:].sum(axis=1)
min_text['NonToxics']=(sum_toxic_scores==0)
```

```python
labels_sum = min_text.iloc[:,2:].sum()
```

```python
#Plot a distribution of toxic vs. non_toxic labels
plt.figure(figsize=(10,6))
sns.barplot(labels_sum.index, labels_sum.values)
plt.title('Distribution of toxic vs non_toxic labels')
plt.tight_layout()
```
